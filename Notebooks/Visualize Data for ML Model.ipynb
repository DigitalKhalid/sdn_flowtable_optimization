```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
data_file = 'ml_dataset_1.csv'
data = pd.read_csv(data_file)
```

```python
data['packet_interval'].value_counts()
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x = data['source_mac_int'], data = data, palette = 'hls')
plt.show()
```

```python
sns.histplot(data['packet_interval'], kde = True, bins = 20, palette = 'hls')
```

```python
sns.distplot(data['packet_interval'], kde = True, bins = 20)
```

```python
sns.boxplot(data['packet_interval'], palette = 'hls')
```

```python
sns.lineplot(x = data['source_mac_int'], y = data['packet_interval'], palette = 'hls')
plt.xticks(rotation=90)
```

```python
sns.barplot(x = data['source_mac_int'], y = data['packet_interval'], palette = 'hls')
```

```python
sns.scatterplot(x = data['source_mac_int'], y = data['packet_interval'], palette = 'hls')
```

```python
sns.pairplot(data, palette = 'hls')
```
